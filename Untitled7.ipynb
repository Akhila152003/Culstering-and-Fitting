```python
def preprocess_data(data):
    # Select only numeric columns for filling missing values
    numeric_data = data.select_dtypes(include=np.number)

    # Fill missing values with mean for numeric columns only
    data[numeric_data.columns] = numeric_data.fillna(numeric_data.mean())

    # Ensure columns are in the correct type (This part is likely okay)
    data['GDP (USD)'] = pd.to_numeric(data['GDP (USD)'], errors='coerce')
    data['Population'] = pd.to_numeric(data['Population'], errors='coerce')
    data['Life Expectancy'] = pd.to_numeric(data['Life Expectancy'], errors='coerce')
    data['Unemployment Rate (%)'] = pd.to_numeric(data['Unemployment Rate (%)'], errors='coerce')
    data['CO2 Emissions (metric tons per capita)'] = pd.to_numeric(data['CO2 Emissions (metric tons per capita)'], errors='coerce')
    data['Access to Electricity (%)'] = pd.to_numeric(data['Access to Electricity (%)'], errors='coerce')

    return data
    print("Data preprocessing completed.")

# Preprocess the data
data = preprocess_data(data)
```

```python
def kmeans_clustering(data, n_clusters=3):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    clusters = kmeans.fit_predict(data[['GDP (USD)', 'Population', 'CO2 Emissions (metric tons per capita)']])
    return clusters, kmeans

# Perform K-means clustering
clusters, kmeans = kmeans_clustering(data)

# Plot scatter plot for clusters
def plot_scatter(data, clusters):
    plt.figure(figsize=(8, 6))
    plt.scatter(data['Population'], data['GDP (USD)'], c=clusters, cmap='viridis', marker='o')
    plt.title("Scatter Plot of Clusters")
    plt.xlabel('Population')
    plt.ylabel('GDP (USD)')
    plt.colorbar(label='Cluster')
    plt.show()
    print(f"K-means clustering performed with {len(set(clusters))} unique clusters.")

plot_scatter(data, clusters)
```

```python
def line_fitting(data):
    X = data[['GDP (USD)']].values  # Replace with your feature
    y = data['Life Expectancy'].values  # Replace with your target variable
    model = LinearRegression()
    model.fit(X, y)
    return model

model = line_fitting(data)

# Plot regression line
def plot_regression_line(data, model):
    plt.figure(figsize=(8, 6))
    plt.scatter(data['GDP (USD)'], data['Life Expectancy'], color='blue', label='Data Points')
    plt.plot(data['GDP (USD)'], model.predict(data[['GDP (USD)']]), color='red', label='Regression Line')
    plt.title("Linear Regression Line Fitting")
    plt.xlabel('GDP (USD)')
    plt.ylabel('Life Expectancy')
    plt.legend()
    plt.show()
    print("Linear regression model fitted. The equation is: y = {:.2f}x + {:.2f}".format(model.coef_[0], model.intercept_))

plot_regression_line(data, model)
```

```python
def plot_elbow(data):
    inertia = []
    for k in range(1, 11):
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(data[['GDP (USD)', 'Population']])  # Replace with your features
        inertia.append(kmeans.inertia_)

    plt.figure(figsize=(8, 6))
    plt.plot(range(1, 11), inertia, marker='o', color='b')
    plt.title("Elbow Method for Optimal K")
    plt.xlabel("Number of Clusters")
    plt.ylabel("Inertia")
    plt.show()
    print("Elbow method completed. Inertia plotted for k=1 to k=10.")

plot_elbow(data)
```

```python
def plot_silhouette(data):
    silhouette_scores = []
    for k in range(2, 11):
        kmeans = KMeans(n_clusters=k, random_state=42)
        clusters = kmeans.fit_predict(data[['GDP (USD)', 'Population']])  # Replace with actual features
        silhouette_scores.append(silhouette_score(data[['GDP (USD)', 'Population']], clusters))

    plt.figure(figsize=(8, 6))
    plt.plot(range(2, 11), silhouette_scores, marker='o', color='green')
    plt.title("Silhouette Score for Optimal K")
    plt.xlabel("Number of Clusters")
    plt.ylabel("Silhouette Score")
    plt.show()
    print(f"Silhouette scores for k=2 to k=10 plotted.")

plot_silhouette(data)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
data['Life Expectancy Category'] = data['Life Expectancy'].apply(lambda x: 1 if x >= 70 else 0)

X = data[['GDP (USD)', 'Population']]  # Features
y = data['Life Expectancy Category']  # Target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train a classifier (Logistic Regression as an example)
clf = LogisticRegression()
clf.fit(X_train, y_train)

# Predict and plot confusion matrix
y_pred = clf.predict(X_test)
cm = confusion_matrix(y_test, y_pred)

# Plot confusion matrix
def plot_confusion_matrix(cm):
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title("Confusion Matrix")
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()
    print("Confusion matrix plotted for Logistic Regression.")

plot_confusion_matrix(cm)
```
